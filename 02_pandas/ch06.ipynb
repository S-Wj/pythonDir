```python
import numpy as np
```

```python
data = np.random.randint(-100,100,50).cumsum()
data
```

```python
import matplotlib.pyplot as plt
```

```python
# 아래의 명령어 코드 실행 안하면 오류남 주석 풀면 정상적으로 실행됨
# plt.rcParams['font.family'] = 'Nanum Pen Script'
plt.plot(range(50),data,'r')

plt.title('시간별 가격 추이')
plt.ylabel('주식가격')
plt.xlabel('시간(분)')
```
