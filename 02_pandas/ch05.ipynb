```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/concat_1.csv')
df2 = pd.read_csv('data/concat_2.csv')
df3 = pd.read_csv('data/concat_3.csv')
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# 목록으로 묶어줌
# 인덱스가 같은 것들끼리 묶어줌.
pd.concat([df1, df2, df3], axis=0) 
```

```python
new = pd.Series(['n1','n2','n3','n4'])
```

```python
pd.concat([df1,new])
```

```python
# 대괄호 2개 했을때
new_df = pd.DataFrame([['n1','n2','n3','n4']],columns=['A','B','C','D'])

# 대괄호 1개만 했을때는 1행만 나옴
# pd.DataFrame(['n1','n2','n3','n4'])
```

```python
df1.columns
```

```python
# 인덱스가 같은거
pd.concat([df1,new_df])
```

```python
df1.append(new_df)
```

```python
pd.concat([df1, df2, df3], axis=0, ignore_index=True)
```

```python
pd.concat([df1, new], join='inner')
# 이러면 제대로 안나오는게 정상임
```
