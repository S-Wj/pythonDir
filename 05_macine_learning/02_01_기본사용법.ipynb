# 머신러닝
컴퓨터가 스스로 학습하여 인공지능의 성능 향상 시키는 기술 방법

# 딥러닝
인간의 뉴런과 비슷한 인공신경망 방식으로 정보를 처리



## 지도학습 
: 문제와 정답을 모두 알려주고 공부시키는 방법

## 비지도학습
: 답을 알려주지 않고 학습하는 방법 
- 정답이 없음
- 입력 데이터만 있음

## 강화학습
: 더 나은 결과를 도출하기 위해 결과값끼리 경쟁하여 보상을주고 더 나은 결과를 탐색


- 시험공부 기간 : 회귀 / pass, fail : 분류

### 회귀
: 결과값을 예측하는 것.

#### 선형 회귀
: 데이터의 특성과 상관관계 등을 파악하는 것.
주어진 데이터를 통해서 일차 방정식을 수정해나간다.
학습을 통해서 가장 합리적인(근사치인) 값을 찾는다.

- 독립변수 : 독립적인 변수
- 종속변수 : 독립변수에 의해 영향을 받는 종속변수

머신러닝이 여기서 가중치와 절편 값에 대해서 학습해 나간다.
와중에 절편 값은 가중치 계산보다 중요하진 않아서 기본 값을 0으로 주고 시작하기도 한다.







- 오차 : `실제 값`과 `예측 값`의 차이를 구한 결과가 `오차 값`이다.
`예측 값`은 `실제 값`과 +, - 만큼 차이 날 수 있기 때문에
`오차 값` 의 절대값을 찾기 위해 제곱의 형태로 만들어서 계산한다.
그 이후에 차이 값들의 합의 평균을 `오차`라고 본다.





#### 경사하강법
: 손실함수 E(W, b)가 최소값을 갖도록 (W, b)를 구하는 것이 선형모델의 목적이다.
이것을 경사하강법이라 본다.









##### W 값
W 값이 0인 값을 찾아야 한다. 그러기 위해서 오차값들의 제곱에서 미분을 통해 W 값의 기울기가 0인 부분을 구한다.
W 값이 0이 아닐때는 0으로 맞추기위해 W 값을 수정(이동)해줘야 하는데. 
그걸 이동하는 정도를 학습률이라고 한다. 이 학습률은 인간이 지정해줘야하는 부분이다.

### 분류 
: 결과를 나누는 것. 

#### 지도학습(분류)

- 의사결정 트리
이런 방법으로도 학습 할 수 있다. 단일 형태로 쓰지 않고 앙상블 형태로 사용하면 성능이 나쁘지 않다.

- k-최근접 이웃 분류기

### 클러스터일
: 몇 가지로 분류할지 정하는 것..

#### 비지도학습(군집)

- 평균군집

# 사이킷런
파이썬 머신러닝 라이브러리이다.

```python
import sklearn
```

```python
sklearn.__version__
```

```python
from sklearn.datasets import load_iris
# 의사결정 트리 사용
from sklearn.tree import DecisionTreeClassifier
# 데이터 분리하는데 사용하는 모델
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
iris = load_iris(as_frame=True)

#어레이 처럼 들어가 있음
```

```python
iris.data.head(2)
```

```python
iris.target
```

```python
iris_df = iris.data
```

```python
# iris_df['label'] = iris.target
```

```python
iris_df.head(2)
```

```python
iris.target_names[0]
```

```python
# 데이터 분리
#   8   :   2  /   8  :   2
X_train,X_test,y_train,y_test = train_test_split(iris.data,iris.target,test_size=0.2,random_state=11)
```

```python
# 랜덤 스테이트 값만 고정 
dt_clf = DecisionTreeClassifier(random_state=11)
```

```python
# 데이터 학습 (정보/결과)
dt_clf.fit(X_train,y_train)
```

```python
# 테스트 데이터 (예측할 정보)
pred = dt_clf.predict(X_test)
```

```python
list(zip(pred,y_test))
```

```python
# 예측 정확도 출력하기
from sklearn.metrics import accuracy_score
```

```python
# 예측 결과 출력
accuracy_score(y_test,pred)

# 예측값과 결과가 100퍼센트 일치함.
```

```python
# 테스트 데이터는 예측 데이터랑 같은 형식으로 넣어줘야 함.
dt_clf.predict([[4.67,3.9,2,0.4]])

# 예측결과 0이니까 setousa 라고 예측함.
```

```python
# 사이킷런은 자체 저장 메서드가 없으니까 파일을 저장하기 위해서 피클 모듈을 사용한다.
# 파일 객체로 사용하고 저장하기 위해 피클 임포트
import pickle
```

```python
# 파일 저장하기
f = open('model','wb')
# 파일로 읽을 때 dump 사용
pickle.dump(dt_clf,f)
```

```python
f.close()
```
