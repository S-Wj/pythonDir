```python
# ParserError : 구분자가 틀려서 에러가 남. csv는 쉼표로 구분되는데 안맞았다는 것.
# `sep='\s+'` 그래서 이렇게 구분 방법 선언해줌. 공백으로 구분하라는 선언
feature_name_df = pd.read_csv('human_activity/features.txt',
                            sep='\s+',
                            header=None,
                            names=['column_index','column_name'])
```

```python
# 중복된 피처명 때문에 생기는 오류 잡기 위해서 조정이 필요함.
# 언더바로 일렬번호를 붙이는 작업 수행
```

```python
feature_dup_df = feature_name_df.groupby('column_name').count()
feature_dup_df.head(2)
```

```python
# 중복 되는거 조회.
feature_dup_df[feature_dup_df['column_index']>1].count()
```

- `.cumcount()`



```python
# 중복 되는 것들 일렬 번호 매겨주는 함수 생성
def get_new_df(old_df):
    dup_df = pd.DataFrame(data=old_df.groupby('column_name').cumcount(),columns=['dup_cnt'])
    dup_df = dup_df.reset_index()
    # 같은 인덱스끼리 연결해서 뒤에 dup_cnt 추가해줌.
    new_df = pd.merge(old_df.reset_index(),dup_df,how='outer')
    # 가져올 컬럼 두개 선택함. 
    # 행을 기준으로 묶어서 들어오는 경우는 axis 기본값(0) 일 경우.
    # axis 값 1로 주면 행을 기준으로 들어와서 하나씩 들어와짐.
    # 이렇게 해야 이 컬럼 값의 뒤에 `_`를 연결시켜서 값을 넣어줌.
    # 파이썬에선 데이터 타입이 같아야 연산이 가능함.
    new_df['column_name'] = new_df[['column_name','dup_cnt']].apply(lambda X:X[0]+'_'+str(X[1]) if X[1]>0 else X[0], axis=1)
    new_df.drop(columns=['index'],inplace=True)
    
    return new_df
```

```python
df = get_new_df(feature_name_df)
df.head(2)
```

```python
df[df['dup_cnt']>0]
```

```python
def get_human_dataset():
    feature_name_df = pd.read_csv('human_activity/features.txt',
                            sep='\s+',
                            header=None,
                            names=['column_index','column_name'])
    name_df = get_new_df(feature_name_df)
    feature_name = name_df.iloc[:,1].values.tolist()
    X_train = pd.read_csv('human_activity/train/X_train.txt',sep='\s+',names=feature_name)
    X_test = pd.read_csv('human_activity/test/X_test.txt',sep='\s+',names=feature_name)
    y_train = pd.read_csv('human_activity/train/y_train.txt',sep='\s+',names=['action'])
    y_test = pd.read_csv('human_activity/test/y_test.txt',sep='\s+',names=['action'])
    return X_train,X_test,y_train,y_test
```

```python
X_train,X_test,y_train,y_test = get_human_dataset()
```

```python
X_train.head(2)
```

```python
X_train.info()
```

```python
# 요소 값들을 뽑고 해당 값들이 몇번 나오는지 확인
y_train['action'].value_counts()

# 각각 나름 균일하게 데이터 수가 존재함.
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
dt_clf = DecisionTreeClassifier(random_state=156)
dt_clf.fit(X_train,y_train)
pred = dt_clf.predict(X_test)
# 정확도 조회
accuracy_score(y_test,pred)
```

```python
# 어떤 파라미터들을 기준으로 처리했는지 확인가능
dt_clf.get_params()
```

```python
# 하이퍼 파라미터 튜닝 작업 실습
```

```python
# GridSearchCV는 머신러닝에서 모델의 성능향상을 위해 쓰이는 기법중 하나입니다.
# 사용자가 직접 모델의 하이퍼 파라미터의 값을 리스트로 입력하면 값에 대한 경우의 수마다 예측 성능을 측정 평가하여 비교하면서 최적의 하이퍼 파라미터 값을 찾는 과정을 진행합니다.
# 시간이 오래걸립니다.
from sklearn.model_selection import GridSearchCV
```

```python
# 딕셔너리 생성
# 파라미터 값 지정하여 조건의 범위를 지정해준다.
params = {
    'max_depth' : [6,8,10,12,16,20,24],
    'min_samples_split' : [16],
}

# `,verbose=` : 진행 사항이 나오게끔 설정해줌.
grid_cv = GridSearchCV(dt_clf,params,scoring='accuracy',cv=5,verbose=1)

# folds 값이 안나오면 안됐다는 뜻.

# 코드 돌린 시간 나타냄
# %%time
grid_cv.fit(X_train,y_train)
```

```python
grid_cv.best_score_
```

```python
grid_cv.best_params_
```

```python
cv_result = pd.DataFrame(grid_cv.cv_results_)
```

```python
cv_result[['param_max_depth','mean_test_score']]

# 파라미터 설정 값 마다 점수가 어떤지 df로 확인해봄
```

```python
# for문으로 깊이 설정에 따른 정확도를 확인해본다.
# 과적합을 방지하기 위해 이런 방식들로 조회해서 확인 해본다.
max_depth = [6,8,12,16,20,24]
for depth in max_depth:
    dt_clf = DecisionTreeClassifier(max_depth=depth,min_samples_split=16,random_state=156)
    dt_clf.fit(X_train,y_train)
    pred = dt_clf.predict(X_test)
    accurary = accuracy_score(y_test,pred)
    print(f'max_depth:{depth} 정확도:{accurary:.4f}')
```

```python
%%time
# 파라미터값 바꿔서 다른 경우의 수도 확인해보자.

# 파라미터 값 지정하여 조건의 범위를 지정해준다.
params = {
    'max_depth' : [8],
    'min_samples_split' : [8,12,16,20,24],
}
grid_cv = GridSearchCV(dt_clf,params,scoring='accuracy',cv=5,verbose=1)

grid_cv.fit(X_train,y_train)
```

```python
# 가장 좋은 파라미터 조합 값 확인
grid_cv.best_params_
```

```python
# 가장 좋은 파라미터 조합의 점수 확인
grid_cv.best_score_
```

```python
# 가장 좋은 경우의 
pred = grid_cv.best_estimator_.predict(X_test)
accuracy_score(y_test,pred)
```

```python
# 어느 어느 컬럼에 중요도(규칙 생성 영향력이 높은 컬럼)가 어떻다 알려줌.
data = pd.Series(grid_cv.best_estimator_.feature_importances_,index=X_train.columns)
```

```python
# 중요도 값이 높은 순서대로 피처들을 확인 (10개만)
top10 = data.sort_values(ascending=False)[:10]
```
