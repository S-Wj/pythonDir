# 지도에 데이터 시각화하기

```python
import folium
```

#### [folium 사용법 기본 안내 홈페이지](https://python-visualization.github.io/folium/quickstart.html)
#### [지도 json 타입 정보 사이트](https://geojson.io/#map=2/0/20)
---
- 위도, 경도 값은 튜플이라서 로케이션 생략해도 된다.
- `,zoom_start=16` 숫자 값으로 확대 비율 정해준다.
- `,tiles='Stamen Terrain'` 값으로 지도 스타일 바꿔줄 수 있다.
---
- 확대 축소하면서 다양하게 사용하기 위해 이렇게 저장해줘야 좋다. # m.save('map.html') 그럼 html파일로 지도가 저장된다.
---
- `,popup='문래역'` 클릭하면 나오는 정보
- `,tooltip='여기는 어디!!'` 마우스 커서 올리면 뜨는 정보
- `,icon=folium.Icon(icon="cloud")` 북마크 아이콘 그림 바꾸는 정보
---
- `radius=20` 반지름 크기
- `,color='red'` 반지름 테두리 색깔
- `,fill_color='orange'` 반지름 내부 색깔
- `,fill_opacity=0.5` 

```python
# 위도 경도 값 넣어서 위치 지도 호출하는 과정.
m = folium.Map((37.5169837,126.8993953),zoom_start=16) #tiles='Stamen Watercolor'

# html 파일로 지도 형식 저장하는 방법.
# 확대 축소하면서 다양하게 사용하기 위해 이렇게 저장해줘야 좋다.
# m.save('map.html')

# 지도에 북마크 남기는 과정.
folium.Marker((37.5169837,126.8993953),
              popup='문래역',
              tooltip='여기는 어디!!',
              icon=folium.Icon(icon="cloud"),).add_to(m)

# 지도에 동그라미 마크 남기는 과정
folium.CircleMarker((37.5169837,126.8993953),
                    radius=20,
                    color='red',
                    fill_color='orange',
                    fill_opacity=0.5).add_to(m)

m
```
